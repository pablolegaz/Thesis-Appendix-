# Generate thesis tables

This notebook exports compact CSV and LaTeX-style tables from the raw result files.

```python

from pathlib import Path
import pandas as pd
ROOT = Path('..').resolve()
RAW = ROOT / 'results' / 'raw'
OUT = ROOT / 'results' / 'processed'
OUT.mkdir(parents=True, exist_ok=True)
files = list(RAW.glob('bfm_*.csv'))
frames=[]
for f in files:
    df=pd.read_csv(f)
    if 'width' not in df.columns:
        df['width']=df['plausibility']-df['belief']
    frames.append(df)
all_results=pd.concat(frames, ignore_index=True)
all_results.to_csv(OUT/'all_bfm_results_combined.csv', index=False)
print('Wrote', OUT/'all_bfm_results_combined.csv')
```

```python

conflict=(all_results.groupby(['scenario_id','level'], as_index=False)['empty_set_mass'].first())
conflict.to_csv(OUT/'conflict_table_long.csv', index=False)
conflict.pivot(index='scenario_id', columns='level', values='empty_set_mass').to_csv(OUT/'conflict_table.csv')
print(conflict)
```
